```python
class parent:
    def __init__(self,ward,father,usn):
        self.ward = ward
        self.father = father
        self.usn = usn
    def result(self,percentage):
        return f'Mr. {self.father} your ward {self.ward} bearing {self.usn} secured {percentage} in the current semester'
    
class college(parent):
    def __init__(self,ward,father,usn,PRINCIPAL,HOD,DEAN):
        parent._init__(self,ward,father,usn)
        self.PRINCIPAL = PRINCIPAL
        self.HOD = HOD
        self.DEAN = DEAN
    def info(self):
        return f'Mr. {self.father} you must meet pricipal {self.PRINCIPAL} and bearing {self.HOD} on or before this month end'
```

```python
information3 = college('Dr suresh','GSS','RAO')
print(information3.info())
```
